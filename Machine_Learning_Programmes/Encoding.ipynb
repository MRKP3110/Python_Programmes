```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
```

```python
df=pd.read_csv("Datasets/train.csv")
```

```python
df
```

store into new variable :

```python
o=OrdinalEncoder(categories=[["male","female"]]) 
```

convert string into numeric :

```python
o.fit_transform(df["sex"].values.reshape(-1,1)) 
```

```python
o.fit_transform(df[["sex"]])  
```

direct store into actual data :

```python
df["sex"]=o.fit_transform(df[["sex"]]) 
```

```python
df
```

conver float into int :

```python
df["sex"]=df["sex"].astype(int)  
```

```python
df
```

how to do with pandas :

```python
pd.get_dummies(df,columns=["embarked"]) 
```

```python
ohe=OneHotEncoder(drop="first")
```

```python
ohe.fit_transform(df[["embarked"]]).toarray()
```

NOTE:  last column is for null values.

```python
ohe=OneHotEncoder(drop="first",sparse_output=False)  
```

NOTE:  if use sparse_output function with False value then we don't need to use toarray function.

```python
ohe.fit_transform(df[["embarked"]])
```

remove or fill null values :

```python
df["embarked"]=df["embarked"].fillna(df["embarked"].mode()[0]) 
```

```python
ohe.fit_transform(df[["embarked"]])  #conver string in numeric
```
