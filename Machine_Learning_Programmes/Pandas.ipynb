```python
pd.concat([df,df1,axis=1])
```

Error Explanation:  You're getting a SyntaxError because you accidentally placed a keyword argument (axis=1) inside a list, which is not valid Python   
                    syntax.

Error Solution: You need to pass the axis=1 argument outside the list, because it is a named parameter to the pd.concat() function—not part of the list 
                of DataFrames.
