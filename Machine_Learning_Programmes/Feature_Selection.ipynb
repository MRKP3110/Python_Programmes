```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest , RFE
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Datasets/train.csv")
```

```python
df
```

```python
df.drop(["name","cabin","ticket"],axis=1 , inplace=True)
```

```python
o = OrdinalEncoder(categories  = [["male","female"]])
```

```python
df["sex"] = o.fit_transform(df[["sex"]])
```

```python
df
```

```python
df["age"]=df["age"].fillna(df["age"].mean())
```

```python
df["embarked"]= df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
ohe = OneHotEncoder(drop="first",sparse_output=False)
```

```python
x = ohe.fit_transform(df[["embarked"]])
```

```python
columns = ohe.get_feature_names_out()
```

```python
df1 = pd.DataFrame(x , columns = columns)
```

```python
pd.concat([df,df1],axis = 1)   # concat is use to mearge to differant dataframe
```

```python
df = pd.concat([df,df1],axis = 1) 
```

```python
df.drop(["embarked"],axis = 1 , inplace=True)
```

```python
df
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(["survived"],axis = 1),df["survived"] , test_size = 0.3 , random_state=32)
```

```python
x_train
```

```python
s = SelectKBest( k = 5)
```

```python
s.fit_transform(x_train , y_train)
```

```python
s.get_feature_names_out()
```

```python
x_train_best = s.fit_transform(x_train , y_train)
x_test_best = s.fit_transform(x_test , y_test)
```

```python
l = LogisticRegression()
```

```python
l.fit(x_train_best , y_train)
```

```python
y_prd = l.predict(x_test_best)
```

```python
accuracy_score(y_test , y_prd)
```

```python
l1 = LogisticRegression()
```

```python
r = RFE(estimator=l1 , n_features_to_select=4)
```

```python
x_train_r = r.fit_transform(x_train_best , y_train)
x_test_r = r.fit_transform(x_test_best , y_test)
```

```python
r.get_feature_names_out()
```

```python
l1.fit(x_train_r, y_train)
```

```python
y_prd1 = l1.predict(x_test_r)
```

```python
accuracy_score(y_test , y_prd1)
```
